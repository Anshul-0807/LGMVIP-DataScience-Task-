```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

def pencil_converter(image_path):

    image = cv2.imread(image_path)
    
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    inverted_image = 260 - gray_image
    
    blurred_image = cv2.GaussianBlur(inverted_image, (25, 25), sigmaX=0, sigmaY=0)
    
    pencil_sketch = 260 - blurred_image
    
    return pencil_sketch

input_image_path = 'rohit_sharma.jpg'

sketch = pencil_converter(input_image_path)

plt.figure(figsize=(20, 12))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(cv2.imread(input_image_path), cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Pencil Sketch')
plt.imshow(sketch, cmap='gray', vmin=0, vmax=255) 
plt.axis('off')

plt.show()
```

```python
input_image_path = 'crowd.jpg'

sketch = pencil_converter(input_image_path)

plt.figure(figsize=(20, 12))
plt.subplot(5, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(cv2.imread(input_image_path), cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Pencil Sketch')
plt.imshow(sketch, cmap='gray', vmin=0, vmax=255)  # Add vmin and vmax to set the color map range
plt.axis('off')

plt.show()
```
